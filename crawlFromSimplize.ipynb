```python
url = 'https://simplize.vn/co-phieu/' + stock + '/lich-su-gia'
```

```python
driver = webdriver.Edge(executable_path='msedgedriver.exe')
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
#soup
```

```python
box = soup.find_all('li', {'tabindex': '0'})
box
```

```python
# find last page index
box = soup.find_all('li', {'tabindex': '0'})
#box
lastIndex = box[-2].find('a').get_text()
lastIndex = int(lastIndex)
lastIndex
```

i from 1 to lastindex, next page next page and scrape

```python
headers = [
    'date', # ngay thang
    'open', # gia mo cua
    'highest', # gia cao nhat
    'lowest', # gia thap nhat
    'close', # gia dong cua
    'change', # gia thay doi
    'change_percentage', # % gia thay doi
    'volume' # khoi luong giao dich
]

df = pandas.DataFrame(columns = headers)
df
```

```python
index = 1
while True:
    soup = BeautifulSoup(driver.page_source, 'lxml')
    rows = soup.find_all('tr', {'class': 'simplize-table-row simplize-table-row-level-0'})
    for row in rows:
        data = crawlOneDay.crawler(row)
        newRow = pandas.DataFrame(data, index=[0])
        df = pandas.concat([df, newRow], ignore_index=True)
    sleep(1)
    # try:
    #     driver.find_element(By.CLASS_NAME, 'simplize-pagination-next').click()
    # except NoSuchElementException:
    #     break
    
    if index < lastIndex:
        # next page
        driver.find_element(By.CLASS_NAME, 'simplize-pagination-next').click()
    else:
        break
    index += 1

df
```
