```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with datetime index and 'close' column
# Example DataFrame creation (replace this with your actual DataFrame)
data = {
    'date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
    'close': [100, 105, 110, 108]
}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Plotting the line chart
plt.figure(figsize=(12, 6))  # Set the figure size (width, height)

# Plotting the 'close' column against the datetime index
#plt.plot(df.index, df['close'], marker='o', linestyle='-', color='b', label='Close Price')
# Using seaborn's lineplot to plot 'close' against the datetime index
sns.lineplot(data=df, x=df.index, y='close', color='g', label='Close Price')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Price (VND)')
plt.title('Stock Close Prices Over Time')

# Displaying grid
plt.grid(True)

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
